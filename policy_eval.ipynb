```python
from pathlib import Path
import csv
import io

MINUTES_PER_DAY = 24 * 60

def load_calls_table(path: str):
    lines = Path(path).read_text().splitlines()
    header_idx = next(i for i, line in enumerate(lines) if line.startswith("index,"))
    reader = csv.DictReader(io.StringIO("\n".join(lines[header_idx:])))
    rows = [row for row in reader if row.get("index")]
    return rows

def average_response_minutes(path: str):
    rows = load_calls_table(path)
    response_days = [float(row["responseDuration"]) for row in rows]
    avg_days = sum(response_days) / len(response_days)
    return {
        "num_calls": len(response_days),
        "avg_response_days": avg_days,
        "avg_response_minutes": avg_days * MINUTES_PER_DAY,
    }

results = {
    "default": average_response_minutes("output/bronx_default/calls.csv"),
    "new_policy": average_response_minutes("output/bronx_new_policy/calls.csv"),
}

results["difference_minutes"] = (
    results["new_policy"]["avg_response_minutes"]
    - results["default"]["avg_response_minutes"]
)

results
```
